# **Arvore de decisão - Melhores abordagens com dados para problemas**

# **Arvore de Decisao - CAUSALIDADE**

```python
from graphviz import Digraph
from google.colab import files
from IPython.display import Image, display

# Cria um novo gráfico direcionado
dot = Digraph()

# Adiciona nós para a árvore de causalidade
dot.node('A', 'Qual é o objetivo da análise causal?')
dot.node('B', 'Identificar relações causais')
dot.node('C', 'Deseja realizar um experimento controlado?')
dot.node('D', 'Sim: Usar A/B Testing\n(Resultados: Comparação direta entre grupos)')
dot.node('E', 'Não: Análise observacional')
dot.node('F', 'Existem variáveis de confusão?')
dot.node('G', 'Sim: Usar Propensity Score Matching\nou Ajuste por Covariáveis\n(Resultados: Efeito isolado)')
dot.node('H', 'Não: Usar Modelos de Regressão\n(Resultados: Estimativas de efeito)')
dot.node('I', 'Explorar relações temporais')
dot.node('J', 'Os dados são de séries temporais?')
dot.node('K', 'Sim: Usar Granger Causality\n(Resultados: Previsibilidade entre séries)')
dot.node('L', 'Não: Análise causal padrão')
dot.node('M', 'Visualizar relações causais')
dot.node('N', 'Deseja usar diagramas causais?')
dot.node('O', 'Sim: Criar DAGs\n(Resultados: Visualização clara de relações)')
dot.node('P', 'Não: Usar outras técnicas de inferência causal.')

# Conexões
dot.edge('A', 'B')
dot.edge('B', 'C')
dot.edge('C', 'D', label='Sim')
dot.edge('C', 'E', label='Não')
dot.edge('E', 'F')
dot.edge('F', 'G', label='Sim')
dot.edge('F', 'H', label='Não')
dot.edge('B', 'I')
dot.edge('I', 'J')
dot.edge('J', 'K', label='Sim')
dot.edge('J', 'L', label='Não')
dot.edge('A', 'M')
dot.edge('M', 'N')
dot.edge('N', 'O', label='Sim')
dot.edge('N', 'P', label='Não')

# Renderiza o gráfico e salva como imagem PNG
dot.render('arvore_decisao_causalidade', format='png', cleanup=True)

# Exibe a imagem gerada
display(Image('arvore_decisao_causalidade.png'))

# Baixa o arquivo gerado
files.download('arvore_decisao_causalidade.png')
```

**Analise Bayesiana x Inferencia Causal?**

```python
import graphviz

# Criando um objeto Digraph
dot = graphviz.Digraph()

# Definindo os nós
dot.node('A', 'Definir o Problema')
dot.node('B', 'O que você quer entender?\n(Estimativa ou Causa e Efeito?)')
dot.node('C1', 'Dados: Incertezas e Priors\n(Análise Bayesiana)')
dot.node('C2', 'Dados: Causa e Efeito\n(Inferência Causal)')
dot.node('D1', 'Coletar Dados Históricos\nou Priors?')
dot.node('D2', 'Coletar Dados Experimentais\nou Observacionais?')
dot.node('E1', 'Exemplo: Estimar a taxa de conversão\n(com base em dados históricos)')
dot.node('E2', 'Exemplo: Avaliar o impacto de uma campanha publicitária\n(ensaios A/B)')
dot.node('F1', 'Definir Modelo Probabilístico\n(e.g., Beta, Binomial)')
dot.node('F2', 'Definir Modelo de Controle\n(e.g., Regressão, Matching)')
dot.node('G1', 'Amostragem (ex: MCMC)\npara Obter Distribuição Posterior')
dot.node('G2', 'Análise Estatística\n(e.g., Regressão Logística)')
dot.node('H1', 'Avaliação: Distribuição Posterior')
dot.node('H2', 'Avaliação: Estimativa de Efeito')
dot.node('I', 'Comunicação dos Resultados')

# Definindo as arestas
dot.edge('A', 'B')
dot.edge('B', 'C1', label='Estimativa')
dot.edge('B', 'C2', label='Causa e Efeito')
dot.edge('C1', 'D1')
dot.edge('C2', 'D2')
dot.edge('D1', 'E1')
dot.edge('D2', 'E2')
dot.edge('E1', 'F1')
dot.edge('E2', 'F2')
dot.edge('F1', 'G1')
dot.edge('F2', 'G2')
dot.edge('G1', 'H1')
dot.edge('G2', 'H2')
dot.edge('H1', 'I')
dot.edge('H2', 'I')

# Renderizando e visualizando a árvore de decisão
dot.render('decision_tree_b_c', format='png', cleanup=True)

# Exibe a imagem gerada
display(Image('decision_tree_b_c.png'))

# Baixa o arquivo gerado
files.download('decision_tree_b_c.png')
```

# **Arvore de Decisao - NLP**

```python
import graphviz

# Criando um objeto Digraph
dot = graphviz.Digraph()

# Definindo os nós
dot.node('A', 'Definir o Problema\n(Tarefa de Classificação?)')
dot.node('B', 'Os dados são rotulados?')
dot.node('C1', 'Aprendizado Não Supervisionado\n(Exploração e Identificação de Padrões)')
dot.node('C2', 'Técnicas de Aprendizado Não Supervisionado')
dot.node('D1', 'Clustering (Agrupamento)')
dot.node('D2', 'K-Means\n(Segmentação de clientes, agrupamento de documentos)')
dot.node('D3', 'Hierarchical Clustering\n(Análise de similaridade entre textos)')
dot.node('D4', 'DBSCAN\n(Detecção de anomalias em dados)')
dot.node('E1', 'Redução de Dimensionalidade')
dot.node('E2', 'PCA\n(Vizualização e pré-processamento)')
dot.node('E3', 't-SNE\n(Vizualização de grupos em dados complexos)')
dot.node('F1', 'Modelagem de Tópicos')
dot.node('F2', 'LDA\n(Análise de texto, resumo de documentos)')
dot.node('G1', 'Tamanho do Conjunto de Dados')
dot.node('H1', 'Menos de 1.000 amostras?')
dot.node('H2', 'Considere Métodos Simples:\n(Bag of Words + Naive Bayes ou Regressão Logística)')
dot.node('I1', 'Texto é complexo?')
dot.node('I2', 'Modelos Complexos:\n- RNN\n- LSTM\n- GRU\n- Transformers (e.g., BERT, GPT)')
dot.node('I3', 'Modelos Robustos:\n- TF-IDF + SVM\n- Random Forests')
dot.node('J1', 'Requisitos de Interpretação?')
dot.node('K1', 'Prefira Modelos Interpretáveis:\n(Regressão Logística ou Árvores de Decisão)')
dot.node('K2', 'Modelos Complexos:\n(Redes Neurais ou Transformers)')

# Definindo as arestas
dot.edge('A', 'B')
dot.edge('B', 'C1', label='Não')
dot.edge('C1', 'C2')
dot.edge('C2', 'D1')
dot.edge('D1', 'D2')
dot.edge('D1', 'D3')
dot.edge('D1', 'D4')
dot.edge('C2', 'E1')
dot.edge('E1', 'E2')
dot.edge('E1', 'E3')
dot.edge('C2', 'F1')
dot.edge('F1', 'F2')
dot.edge('B', 'G1', label='Sim')
dot.edge('G1', 'H1')
dot.edge('H1', 'H2', label='Sim')
dot.edge('H1', 'I1', label='Não')
dot.edge('I1', 'I2', label='Sim')
dot.edge('I1', 'I3', label='Não')
dot.edge('I2', 'J1')
dot.edge('J1', 'K1', label='Sim')
dot.edge('J1', 'K2', label='Não')

# Renderizando e visualizando a árvore de decisão
dot.render('decision_tree_nlp', format='png', cleanup=True)

display(Image('decision_tree_nlp.png'))

# Baixa o arquivo gerado
files.download('decision_tree.png')
```
